```python
"""
Overall Discussion

First of all, to solve this probelm easily, i assumed that this rectangle is in the graph,
so that i can search and use specific information, such as positions and size of area, easily.
Thus, i made a class 'Towel' with the height, width, starting position x and starting position y
(i.e. towel(1,2,3,4)). In other words, i put them into a graph by giving them integer corners.
i.e) (1,1),(1,3),(3,1),(3,3)
"""
```

```python
"""
It was really hard to find overlapping area with only using dots(corners).
If two rectangles just reach or touch one line or one dots each other, it will regard this as overlapped.
In worst case, if there is overlapping 1 by 1 square,
and if i just remove that same dots, i will totally lose 4 corners that will change the result. 
It has same problem even in the opposite side.
If i just leave them, it is really hard to determine it is an empty space or not."""
```

```python
"""So, i came up with the idea 'Tile'. Basically, Tile is a 1 by 1 square, and i put them into a dictionary
looks like "tile['(x,y),(x+1,y+1)'] = size of area".
If the towel candidate comes in, i divide them into 1 by 1 square by using attribute tile().
Even the key of tile seems a little bit long, It was much easier to find overlapping area and skimming it,
because i can simply remove overlapped tiles and don't care about any other complicated things,
such as overlapping lines, 1 by 1 square and horrible dots."""
```

```python
"""The core of this problem is searching biggest rectangle from the area that has been trimmed.
I solved that problem with the function "reshape".
First, i cut all overlapping area with iteration by checking all the footprints.
i just pop tile, if history already has the same tile(if they already have same key of dictionary tile)."""
```

```python
"""After popping tiles, I did another 2 interations to search the biggest rectangle.
In first iteration, I set every tiles as starting point, and from that starting point, 
by moving 1 tile right and right(if the right next tile exists), i looked up the longest width.
Again i go up to see if upper line has the same width with the starting line.
In the second iteration, i did opposite way. searching longest height by going up first and going right to find longest width.
I append them into a dictionary(result[width*height] = (width,height,i[0],i[1])).
Lastly, the function returns the biggest rectangle.

Finally function "online" integrate all the functions and returns final towel.

Overall, it was effective for me to find biggest rectangle becuase i used tiles.
But i lost time complexity because of that. I checked every tile to get the biggest rectangle.
Even i got the result accurately but it's a bit slow.
Furthermore, i used 'sleep' module to make plotting as video. It also take some efficiency and make is slower
However more than that i think i got most intuitive data and readable code because of that lost."""
```
